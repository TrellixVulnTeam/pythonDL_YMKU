```python
from mxnet import nd
import numpy as np
import matplotlib.pylab as plt
```

```python
def func1(x):
    return np.sum(x**2)
```

```python
def func2(x):
    return np.sum((x-1)**2)
```

```python
def diff(f, x):
    h = 1e-4
    return (f(x+h) - f(x-h)) / (2*h)
```

```python
def grad(f, x):
    h = 1e-4
    grad = np.zeros_like(x)
    
    for i in range(x.size):
        tmp = x[i]
        x[i] = tmp+h
        fxh1 = f(x)
        
        x[i] = tmp-h
        fxh2 = f(x)
        
        grad[i] = (fxh1 - fxh2) / (2*h)
        x[i] = tmp
    return grad
```

```python
x = np.arange(0.0, 20.0, 0.1)
y = func1(x)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.plot(x,y)
plt.show()
```

```python
grad(func1, np.array([1,2,3,4]))
```

```python
def grad_desc(f, init_x, lr=0.01, step_num=100):
    x = init_x
    X,Y = [],[]
    for i in range(step_num):
        X.append(x[0])
        Y.append(x[1])
        x -= lr* grad(f, x)
    plt.scatter(X,Y)
    plt.show()
    return x
```

```python
initx = np.array([-3.0, 4.0])
grad_desc(func2, init_x = initx, lr=0.01, step_num=100)
```
